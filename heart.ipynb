```python
data = pd.read_csv('synthetic_heart_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# univariate anaylsis -- plotting the distribution of categorical variables
categorical_features = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=feature, data=data)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# plotting the distribution of numerical variables
numerical_features = ['Age', 'RestingBP', 'Cholesterol', 'Oldpeak', 'MaxHR']

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[feature], bins=30, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# the machine only understands numerical values so turn all non-numerical columns into numnerical values.

data['Sex'] = data['Sex'].astype('category')
data['Sex'] = data['Sex'].cat.codes

data['ChestPainType'] = data['ChestPainType'].astype('category')
data['ChestPainType'] = data['ChestPainType'].cat.codes

data['RestingECG'] = data['RestingECG'].astype('category')
data['RestingECG'] = data['RestingECG'].cat.codes

data['ExerciseAngina'] = data['ExerciseAngina'].astype('category')
data['ExerciseAngina'] = data['ExerciseAngina'].cat.codes

data['ST_Slope'] = data['ST_Slope'].astype('category')
data['ST_Slope'] = data['ST_Slope'].cat.codes

data
```

```python
# multivariate analysis -- correlation matrix
corr_matrix = data.corr()

# plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# target variable analysis -- box plot to understand outliers and the spread of numerical values

for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='HeartDisease', y=feature, data=data)
    plt.title(f'{feature} by HeartDisease')
    plt.show()
```

```python
# assign X and y values

X = data.drop(columns = 'HeartDisease')
y = data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 20)
# using 70% of the data to train and 30% of the data to test
```

```python
# using a standard scaler to ensure that all the data is equally scaled to prevent biases

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, precision_recall_curve, average_precision_score
```

```python
# binary classification experimentation space
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(max_iter=2000),
}

results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='roc_auc')
    results[name] = scores
    print(f'{name}: Mean ROC AUC = {scores.mean():.2f}, Std = {scores.std():.2f}')
```

```python
# display results
results_df = pd.DataFrame(results).T
results_df.columns = ['Fold1', 'Fold2', 'Fold3', 'Fold4', 'Fold5']
print(results_df)
```

```python
# training a logistic regression model as the primary model
model = LogisticRegression(max_iter = 5000).fit(X_train_scaled, y_train)
```

```python
model.predict(X_train_scaled)
```

```python
y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, model.predict_proba(X_test_scaled)[:, 1])

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
print(f'ROC AUC: {auc:.2f}')
```

```python
model.predict(X_test_scaled)
```

```python
# precision-recall curve
y_scores = model.predict_proba(X_test_scaled)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
average_precision = average_precision_score(y_test, y_scores)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Logistic Regression')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AP={average_precision:.2f})')
plt.legend()
plt.show()
```

```python
# find the best threshold
f1_scores = 2 * (precision * recall) / (precision + recall)
best_threshold_index = np.argmax(f1_scores)
best_threshold = thresholds[best_threshold_index]

print(f'Best Threshold: {best_threshold}')
```

```python
# feature importance from random forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train_scaled, y_train)
feature_importances = rf_model.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# display the most important features
print(importance_df.head(10))
```

```python
# plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df.head(10))
plt.title('Top 10 Important Features')
plt.show()
```

```python
# hyperparameter tuning for random forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train_scaled, y_train)
print(f'Best Parameters: {grid_search.best_params_}')
print(f'Best ROC AUC Score: {grid_search.best_score_:.2f}')
```
