```python
import pandas as pd
```

```python
data = pd.read_csv('heart.csv')
data.head()
```

```python
# the machine only understands numerical values so turn all non-numerical columns into numnerical values.

data['Sex'] = data['Sex'].astype('category')
data['Sex'] = data['Sex'].cat.codes

data['ChestPainType'] = data['ChestPainType'].astype('category')
data['ChestPainType'] = data['ChestPainType'].cat.codes

data['RestingECG'] = data['RestingECG'].astype('category')
data['RestingECG'] = data['RestingECG'].cat.codes

data['ExerciseAngina'] = data['ExerciseAngina'].astype('category')
data['ExerciseAngina'] = data['ExerciseAngina'].cat.codes

data['ST_Slope'] = data['ST_Slope'].astype('category')
data['ST_Slope'] = data['ST_Slope'].cat.codes

data
```

```python
# visualizing data just to understand it a little better
import seaborn as sns
sns.pairplot(data)

### use bargraphs for age height etc to see what has the highest correlations
### use missingo to see for null values instead of the boring way???
```

```python
data.isnull().sum()
```

```python
#yayy no null values!

#now assign which columns are X and y.

X = data.drop(columns = 'HeartDisease')
y = data['HeartDisease']
```

```python
X
```

```python
y
```

```python
# now that the data is all prepped, its time to train, test, split.

from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 20) 
# using 70% of the data to train and 30% of the data to test
```

```python
# training data
X_train
```

```python
# testing data
X_test
```

```python
# using a standard scaler to ensure that all the data is equally scaled to prevent biases

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
### we dont want to fit the test data since we dont want the test data to know the mean and standard deviation of the training data because then its basically cheating 
```

```python
X_train_scaled
```

```python
# time to do logistic regression now
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state = 0).fit(X_train_scaled, y_train)
# now the model is done being trained!
```

```python
# check if the model can predict

log_reg.predict(X_train_scaled) # should give me y_train
```

```python
log_reg.score(X_train_scaled, y_train)
```

```python
# model is 85% accurate.
```

```python
# try with the test data

log_reg.score(X_test_scaled, y_test)
```

```python
# the results are roughly 85% accurate!
```

```python
### transform variables to see if they can help strengthen accuracy
```

```python
better_log_reg = LogisticRegression(random_state = 0, C = 0.05, fit_intercept=True).fit(X_train_scaled, y_train)

# C is a regularization parameter that manages extreme values
```

```python
better_log_reg.score(X_train_scaled, y_train)
```

```python
better_log_reg.score(X_test_scaled, y_test)
```

```python
# the C = 0.05 helped improve model performance by a small amount
```
